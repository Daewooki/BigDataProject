# Visualization Tutorial

```python
#pandas import
import pandas as pd
```

```python
#국민건강보험 health data 읽기
data = pd.read_csv("data/health/health_2016.csv")

print(data.shape)
data.head()
```

```python
#seaborn import
import seaborn as sns

%matplotlib inline
```

```python
#matplotlib import

import matplotlib
from IPython.display import set_matplotlib_formats

#Window
matplotlib.rc("font", family="Malgun Gothic")
matplotlib.rc("axes", unicode_minus=False)


set_matplotlib_formats('retina')
```

```python
sns.countplot(data=data, x="성별코드")
```

```python
sns.countplot(data=data, x="시도코드")
```

```python
#pyplot import
import matplotlib.pyplot as plt

plt.figure(figsize=(18,3))
sns.countplot(data=data, x="시도코드")
```

## 두가지 figure 동시에 그리기
* 먼저 set해주고 그림 넣기

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(18,3)
```

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(18,3)

sns.countplot(data=data, x="성별코드", ax=ax1)
sns.countplot(data=data, x="시도코드", ax=ax2)
```

```python
sns.countplot(data=data, x="시도코드", color="#3554E8")
```

```python
# BuGn_r 테마로 그릴 수 있음. 
palette = sns.color_palette("BuGn_r", 10)
sns.countplot(data=data, x="시도코드", palette=palette)
```

## y축에 데이터를 넣어주기

```python
# BuGn_r 테마로 그릴 수 있음. 
palette = sns.color_palette("BuGn_r", 10)
sns.countplot(data=data, y="시도코드", palette=palette)
```

# 분포 그래프 그리기

```python
sns.distplot(data["허리둘레"])
```

```python
sns.distplot(data["허리둘레"], kde=False)
```

```python
sns.distplot(data["허리둘레"], hist=False)
```

```python
# 1 : 흡연 안 함
# 2 : 흡연 했지만 그만둠
# 3 : 흡연 중
data["흡연상태"].value_counts()
```

```python
one = data[data["흡연상태"] == 1]
two = data[data["흡연상태"] == 2]
three = data[data["흡연상태"] == 3]
```

```python
sns.distplot(one["허리둘레"], label="흡연 안 함", hist=False)
sns.distplot(two["허리둘레"], label="흡연 했지만 그만둠", hist=False)
sns.distplot(three["허리둘레"], label="흡연중", hist=False)
```
